# População

`Descrição:` 

`População:` 100x2 matrix with \[0,10) values.

`Useful links:` [Numpy Random Sample](https://docs.scipy.org/doc/numpy-1.15.0/reference/generated/numpy.random.sample.html)

```python
def gen_population(n_rows, n_cols, a, b):
    '''
    This function generates a random populations as a matrix of size n_rows x n_cols and [a,b) values
    
    INPUT: 
    n_rows: number of lines
    n_cols: number of columns
    a: lower interval
    b: higher interval
    
    OUTPUT: 
    np_population: n_rows x n_cols random numpy array with [a, b), b > a values.
    '''

    np_population = (b - a) * random.random_sample((n_rows,n_cols)) + a
    return np_population
```
